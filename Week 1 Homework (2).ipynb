# TASK 1
## 1. Using any finance API (we use Yahoo Finance and Tiingo during sessions) download historical stock prices for Google (symbol ‘GOOG’), for the period 1 Jan 2020 to 1 Jan 2021.


```python
import pandas as pd
import pandas_datareader as pdr


api_key='9418f41a9f45479a17f3da8ea0a63db8f4cd60cb' 

start="2020-1-1"
end="2021-1-1"

df = pdr.tiingo.TiingoDailyReader('GOOG', start=start, end=end, api_key=api_key)


GOOG = df.read()

GOOG
```

## 2. Save these prices to csv file with Pandas


```python
GOOG.to_csv('Google_history_2020_2021.csv',  date_format='%Y-%m-%d') 
```

## 3. Read this csv file to display data saved.


```python
saved_df = pd.read_csv('Google_history_2020_2021.csv', header=0, index_col='date', parse_dates=True)

saved_df
```

## 4. Plot a simple diagram to display closing prices for GOOG stock.

```python
import matplotlib.pyplot as plt
time_close_data = saved_df[['close']]
time_close_data.plot(y='close', kind = 'line', grid = True)
plt.xlabel('Year/Month')
plt.ylabel('Closing Price (Currency in USD)')
plt.title('Closing prices for GOOG stock from Jan 2020 - Jan 2021')
plt.show()
```
